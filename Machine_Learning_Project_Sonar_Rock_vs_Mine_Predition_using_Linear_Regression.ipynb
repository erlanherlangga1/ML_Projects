```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data colletion and data pre processing

```python
sonar_data = pd.read_csv('sonar_data.csv',header=None)
sonar_data.head()
```

```python
# mengetahui jumlah kolom dan baris dari dataframe sonar_data 
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts() # menghitung banyak data pada atribut indeks ke-60 yaitu atribut target/label
```

jika kita memiliki 2 kategori data yang jumlah datanya hampir sama sepertii kasus di atas maka kita akan menghasilkan model yang baik dan juga memiliki tingkat akurasi yang baik.

```python
sonar_data.groupby(60).mean()
```

```python
# untuk membagi attribut predictor ke dalam variabel X dan attribut label/target ke dalam variabel y
# bisa digunakan iloc atau juga dengan drop column
X = sonar_data.iloc[:,:-1]
y = sonar_data.iloc[:,-1]
print(X)
print(y)
```

```python
# cara lain
X = sonar_data.drop(columns=60,axis=1)
y = sonar_data[60]
print(X)
print(y)
```

Split data ke dalam training dan testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify=y, random_state=1)
#stratify digunakan untuk mempertahankan proporsi presentase dari data targe/label yang kita punya pada saat membagi dataset kita kedalam data testing dan data training
#misal pada dataset di atas terdapat data kategorikal [M,R] yang memiliki presentase 55:45
# maka dari data training dan data testing yang kita dapat menemukan atribut target dengan proporsi sebesar 55:45  
```

```python
print(X.shape,X_train.shape, X_test.shape)
```

```python
print(y_test.value_counts(), y_train.value_counts())
```

```python
print(X_train,y_train)
```

Model Training

```python
model = LogisticRegression()
```

```python
# training the logistic regression model with training data
model.fit(X_train,y_train)
```

Model Evaluation
cek tingkat akurasi model yang kita buat

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,y_train)
```

```python
print("Accuracy of training data:", training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,y_test)
print("Accuracy of testing data: ", test_data_accuracy)
```

```python
# menggunakan librari metrics
from sklearn import metrics
y_pred = model.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
